```python
The Kuramoto model can accommodate matrices with negative elements. 

A negative coupling strength implies an anti-phase or repulsive interaction.

Oscillators connected with a negative coupling will tend to synchronize 180 degrees out of phase.
This is a valid extension of the standard Kuramoto model and can lead to interesting dynamics.
```

```python
import numpy as np  
import matplotlib.pyplot as plt  
import networkx as nx  

node_labels = ['LC', 'PRF', 'CNF', 'PPN',  
               'SNr', 'STN', 'GPi', 'GPe',  
               'Str', 'Ctx', 'SNc', 'Th']  

# Define the nodes and the weighted signed adjacency matrix  
nodes = node_labels  
weighted_signed_h = d_pd

# Create a directed graph  
G = nx.DiGraph()  

# Add nodes  
G.add_nodes_from(nodes)  

# Add edges based on the weighted_signed_h matrix  
for i in range(len(nodes)):  
    for j in range(len(nodes)):  
        weight = weighted_signed_h[i][j]  
        if weight != 0:  # only add non-zero weights  
            G.add_edge(nodes[i], nodes[j], weight=weight)  

# Set up the circular layout  
pos = nx.circular_layout(G)  

# Draw the graph  
plt.figure(figsize=(4, 3))  

# Create a list of edges for coloring them  
edges = G.edges(data=True)  
edge_colors = []  

for u, v, data in edges:  
    weight = data['weight']  
    if weight > 0:  
        edge_colors.append('blue')  # Excitatory -> blue  
    else:  
        edge_colors.append('red')  # Inhibitory -> red  

# Draw nodes  
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=1000, font_size=12, font_weight='bold', arrows=True)  
# Draw edges with specified colors  
nx.draw_networkx_edges(G, pos, edge_color=edge_colors)  

# Draw edge labels with weights and set font size  
edge_labels = nx.get_edge_attributes(G, 'weight')  
edge_labels = {k: f"{v:.2f}" for k, v in edge_labels.items()}  # format weights  
# nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=5)  # Adjust font size here  

plt.title("Node Interactions with Weighted Directions (Excitatory in Blue, Inhibitory in Red)")  
plt.axis('off')  # Hide axis  

plt.savefig('results/PD_graph_circle.png')

plt.show()   
```
